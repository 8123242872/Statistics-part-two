```python
#1.What is hypothesis testing in statistics?
"""
Hypothesis testing is a statistical method used to make inferences about a population based on a sample of data. It involves formulating a null hypothesis and an alternative hypothesis, collecting data, and using statistical tests to determine whether the data supports rejecting the null hypothesis in favor of the alternative hypothesis.
"""
```

```python
#2.What is the null hypothesis, and how does it differ from the alternative hypothesis?
"""
The null hypothesis is a statement that assumes there is no significant difference or relationship between variables in a population. The alternative hypothesis is a statement that contradicts the null hypothesis and suggests that there is a significant difference or relationship.

"""
```

```python
#3.What is the significance level in hypothesis testing, and why is it important?
"""
The significance level is the probability of rejecting the null hypothesis when it is actually true. It is denoted by alpha (α). A common significance level is 0.05, which means that there is a 5% chance of rejecting the null hypothesis when it is true. The significance level is important because it determines
"""
```

```python
#4.What does a P-value represent in hypothesis testing?
"""
A P-value is the probability of observing a test statistic as extreme as, or more extreme than, the one observed in the sample, assuming that the null hypothesis is true. It helps determine whether the observed results are statistically significant.
"""
```

```python
#5.How do you interpret the P-value in hypothesis testing?
"""
 Interpreting the P-Value
P-value ≤ α (e.g., 0.05): Reject the null hypothesis. This suggests that the observed data is unlikely under the null hypothesis, indicating statistical significance.

P-value > α: Fail to reject the null hypothesis. This indicates insufficient evidence to support the alternative hypothesis.

For instance, if a study yields a p-value of 0.03 with a significance level (α) of 0.05, the null hypothesis would be rejected, suggesting that the observed effect is statistically significant.


"""
```

```python
#6.What are Type 1 and Type 2 errors in hypothesis testing?
"""
Type 1 Error (False Positive): Rejecting the null hypothesis when it is actually true.
Type 2 Error (False Negative): Failing to reject the null hypothesis when it is actually false.

"""
```

```python
#7.What is the difference between a one-tailed and a two-tailed test in hypothesis testing?
"""
One-Tailed Test: Tests for a specific direction (greater than or less than).
Two-Tailed Test: Tests for any difference, regardless of direction.

"""
```

```python
#8.What is the Z-test, and when is it used in hypothesis testing?
"""
 A Z-test is a statistical method used to determine whether there is a significant difference between a sample mean and a known population mean, or between the means of two independent groups. It is particularly useful when the sample size is large (typically n > 30) and the population standard deviation is known.

  When to Use a Z-Test
Large Sample Size: When the sample size is large (n > 30).
Known Population Standard Deviation: When the population standard deviation is known.
Normal Distribution: When the data is approximately normally distributed.

"""
```

```python
#9.How do you calculate the Z-score, and what does it represent in hypothesis testing?
"""
 Z-Score Calculation
The Z-score is calculated using the formula:

Z = (X̄ − μ) / (σ / √n)

In the context of hypothesis testing, the Z-score is used to determine how far away a sample statistic is from the population parameter under the null hypothesis, in terms of standard errors. This helps assess whether the observed data falls within a specified range of values or if it is statistically significant.

"""
```

```python
#10. What is the T-distribution, and when should it be used instead of the normal distribution?
"""
 The T-distribution is a probability distribution that is similar to the normal distribution but has heavier tails. It is used when the sample size is small (typically n < 30) and the population standard deviation is unknown. The T-distribution accounts for the uncertainty introduced by estimating the population standard deviation from the sample.
 When to Use the T-Distribution
Small Sample Size: When the sample size is small (n < 30).
Unknown Population Standard Deviation: When the population standard deviation is unknown.
Normal Distribution Assumption: When the data is approximately normally distributed.

"""
```

```python
#11.What is the difference between a Z-test and a T-test?
"""
 Z-Test vs. T-Test
Z-Test: Used when the sample size is large (n > 30) and the population standard deviation is known.
T-Test: Used when the sample size is small (n < 30) or the population standard deviation is unknown.

"""
```

```python
#12. What is the T-test, and how is it used in hypothesis testing?
"""
 A T-test is a statistical method used to determine whether there is a significant difference between the means of two groups. It is particularly useful when the sample size is small (typically n < 30) or the population standard deviation is unknown.

  How the T-Test Works
Null Hypothesis (H₀): μ₁ = μ₂ (There is no significant difference between the means of the two groups).
Alternative Hypothesis (H₁): μ₁ ≠ μ₂ (There is a significant difference between the means of the two groups).
Test Statistic: The T-statistic is calculated using the formula:

t = (X̄₁ − X̄₂) / √((s₁²/n₁) + (s₂²/n₂))

Where:
X̄₁ and X̄₂ are the sample means of the two groups.
s₁ and s₂ are the sample standard deviations of the two groups. n₁ and n₂ are the sample sizes of the two groups.
Degrees of Freedom: The degrees of freedom (df) are calculated as df = n₁ + n₂ − 2.
Critical Value: The critical value is determined from the T-distribution table based on the significance level (α) and degrees of freedom.
Decision Rule: If the calculated T-statistic is greater than the critical value, reject the null hypothesis. Otherwise, fail to reject the null hypothesis.


"""
```

```python
#13.What is the relationship between Z-test and T-test in hypothesis testing?
"""
 Z-Test vs. T-Test Relationship
Z-Test: Used when the sample size is large (n > 30) and the population standard deviation is known.
T-Test: Used when the sample size is small (n < 30) or the population standard deviation is unknown. The T-test is similar to the Z-test but accounts for the uncertainty introduced by estimating the population standard deviation from the sample.

"""
```

```python
#14.What is a confidence interval, and how is it used to interpret statistical results?
"""
 A confidence interval is a range of values that is likely to contain the true population parameter with a certain level of confidence. It is used to quantify the uncertainty associated with estimating population parameters from sample data.

  How Confidence Intervals Are Used
Interpretation: A 95% confidence interval means that if we were to repeat the sampling process many times, approximately 95% of the resulting intervals would contain the true population parameter.
Decision Making: Confidence intervals provide a range of plausible values for the population parameter, helping in decision-making processes.
Statistical Significance: Confidence intervals can be used to assess the statistical significance of results. If the confidence interval does not include the null hypothesis value, the result is statistically significant.

"""
```

```python
#15. What is the margin of error, and how does it affect the confidence interval?
"""
 The margin of error is the range of values above and below the point estimate within which the true population parameter is likely to fall. It affects the width of the confidence interval. A smaller margin of error results in a narrower confidence interval, indicating higher precision.
"""
```

```python
#16.How is Bayes' Theorem used in statistics, and what is its significance?
"""
 Bayes' Theorem is a mathematical formula used to update the probability of an event based on new evidence. It is significant because it provides a systematic way to revise beliefs or probabilities when new information becomes available.

"""
```

```python
#17.What is the Chi-square distribution, and when is it used?
"""
 The Chi-square distribution is a probability distribution that arises naturally in the sampling distribution of a test statistic used in hypothesis testing. It is used to test the independence of categorical variables and to assess the goodness of fit of a model to observed data.

"""
```

```python
#18.What is the Chi-square goodness of fit test, and how is it applied?
"""
 The Chi-square goodness of fit test is used to determine whether observed frequencies in a categorical variable match the expected frequencies under a hypothesized distribution. It is applied by comparing observed frequencies with expected frequencies and calculating a test statistic.

"""
```

```python
#19.What is the F-distribution, and when is it used in hypothesis testing?
"""
 The F-distribution is a probability distribution that arises naturally in the sampling distribution of a test statistic used in hypothesis testing, particularly in analysis of variance (ANOVA). It is used to compare variances between groups.

"""
```

```python
#20.What is an ANOVA test, and what are its assumptions?
"""
 ANOVA (Analysis of Variance) is a statistical method used to compare means between three or more groups. Its assumptions include normality of residuals, homogeneity of variances, and independence of observations.
"""
```

```python
#21. What are the different types of ANOVA tests?
"""
 One-Way ANOVA: Compares means between three or more independent groups.
Two-Way ANOVA: Compares means between two independent variables.
Repeated Measures ANOVA: Compares means within the same group at different time points.
Mixed ANOVA: Combines elements of one-way and repeated measures ANOVA.

"""
```

```python
#2.Generate a discrete uniform distribution using Python and plot the probability mass function (PMF).
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import randint

# Parameters
low = 1
high = 7  # Exclusive upper bound, so the distribution is over [1, 2, 3, 4, 5, 6]

# Create a discrete uniform distribution
rv = randint(low, high)

# Values over which to evaluate the PMF
x = np.arange(low, high)

# Compute the PMF
pmf_values = rv.pmf(x)

# Plot the PMF
plt.bar(x, pmf_values, tick_label=x, color='skyblue', edgecolor='black')
plt.title('PMF of Discrete Uniform Distribution (1 to 6)')
plt.xlabel('Outcome')
plt.ylabel('Probability')
plt.grid(True)
plt.show()
```

```python
#3.Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution.
import numpy as np

def bernoulli_pmf(k, p):
    """
    Calculate the Probability Mass Function (PMF) for a Bernoulli distribution.

    Parameters:
    k (int): Outcome (0 or 1)
    p (float): Probability of success (0 <= p <= 1)

    Returns:
    float: PMF value
    """
    if k not in [0, 1]:
        raise ValueError("k must be either 0 or 1.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return p**k * (1 - p)**(1 - k)

# Example usage
p = 0.3  # Probability of success
k_values = [0, 1]
pmf_values = [bernoulli_pmf(k, p) for k in k_values]

print("PMF values:", pmf_values)
```

```python
#9.Write a Python function to calculate Z-scores from a dataset and plot them.
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import zscore

def calculate_and_plot_z_scores(data):
    # Convert data to a pandas DataFrame
    df = pd.DataFrame(data, columns=['Value'])

    # Calculate Z-scores
    df['Z-Score'] = zscore(df['Value'])

    # Plot the Z-scores
    plt.figure(figsize=(10, 6))
    plt.plot(df['Z-Score'], marker='o', linestyle='-', color='b', label='Z-Scores')
    plt.axhline(y=0, color='k', linestyle='--', label='Mean (Z=0)')
    plt.axhline(y=3, color='r', linestyle='--', label='Z=3 (Outlier Threshold)')
    plt.axhline(y=-3, color='r', linestyle='--', label='Z=-3 (Outlier Threshold)')
    plt.title('Z-Scores of Data Points')
    plt.xlabel('Index')
    plt.ylabel('Z-Score')
    plt.legend()
    plt.grid(True)
    plt.show()

    return df
```

```python
#14.Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution.
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_z_score(data_point, mean, std_dev):
    """
    Calculate the Z-score for a given data point.

    Parameters:
    - data_point: The value for which to calculate the Z-score.
    - mean: The mean of the distribution.
    - std_dev: The standard deviation of the distribution.

    Returns:
    - z_score: The Z-score corresponding to the data point.
    """
    z_score = (data_point - mean) / std_dev
    return z_score

def plot_standard_normal_distribution():
    """
    Plot the standard normal distribution curve.
    """
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)
    plt.plot(x, y, label='Standard Normal Distribution')
    plt.title('Standard Normal Distribution Curve')
    plt.xlabel('Z-score')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.legend()
    plt.show()

# Example usage
mean = 0
std_dev = 1
data_point = 2

z_score = calculate_z_score(data_point, mean, std_dev)
print(f"The Z-score for the data point {data_point} is: {z_score}")

# Plot the standard normal distribution
plot_standard_normal_distribution()
```

```python
#16.Create a confidence interval for a dataset using Python and interpret the result.
import numpy as np
import scipy.stats as stats

# Sample data
ages = [23, 25, 30, 22, 28, 35, 40, 29, 31, 27,
        26, 32, 33, 24, 29, 30, 28, 25, 27, 31,
        34, 36, 29, 30, 32, 33, 35, 37, 38, 39]

# Sample size
n = len(ages)

# Sample mean
mean_age = np.mean(ages)

# Sample standard deviation
std_dev = np.std(ages, ddof=1)

# Standard error
se = std_dev / np.sqrt(n)

# Degrees of freedom
df = n - 1

# Critical value for 95% confidence
t_critical = stats.t.ppf(0.975, df)

# Margin of error
margin_of_error = t_critical * se

# Confidence interval
ci_lower = mean_age - margin_of_error
ci_upper = mean_age + margin_of_error

print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#17.Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean.
import numpy as np
import scipy.stats as stats

# Parameters
np.random.seed(0)  # For reproducibility
sample_size = 100
population_mean = 50
population_std = 10
confidence_level = 0.95

# Generate sample data
sample_data = np.random.normal(loc=population_mean, scale=population_std, size=sample_size)

# Calculate sample statistics
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)  # Sample standard deviation (ddof=1 for unbiased estimate)
standard_error = sample_std / np.sqrt(sample_size)

# Determine critical value (Z-score for 95% confidence)
z_critical = stats.norm.ppf((1 + confidence_level) / 2)

# Calculate margin of error
margin_of_error = z_critical * standard_error

# Compute confidence interval
ci_lower = sample_mean - margin_of_error
ci_upper = sample_mean + margin_of_error

# Display results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Critical Value (Z-score): {z_critical:.2f}")
print(f"Margin of Error: {margin_of_error:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
```

```python
#1.Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the result.
import numpy as np
from scipy.stats import norm

# Sample data
sample_mean = 1600
population_mean = 1500
population_std_dev = 100
sample_size = 9

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (population_std_dev / np.sqrt(sample_size))

# Determine the critical value for a 5% significance level (two-tailed test)
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)

# Output the results
print(f"Z-score: {z_score:.2f}")
print(f"Critical value: ±{z_critical:.2f}")

# Decision
if abs(z_score) > z_critical:
    print("Reject the null hypothesis: The sample mean is significantly different from the population mean.")
else:
    print("Fail to reject the null hypothesis: The sample mean is not significantly different from the population mean.")
```

```python
#5.Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing.
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_errors(pop_mean, sample_mean, std_dev, sample_size, alpha=0.05):
    # Calculate the test statistic (Z)
    z_stat = (sample_mean - pop_mean) / (std_dev / np.sqrt(sample_size))

    # Critical z-value for two-tailed test
    z_critical = norm.ppf(1 - alpha/2)

    # Calculate Type I and Type II errors
    type_1_error = alpha
    type_2_error = norm.cdf(z_critical - z_stat) + (1 - norm.cdf(-z_critical - z_stat))

    # Plotting
    x = np.linspace(-4, 4, 1000)
    y = norm.pdf(x)

    plt.plot(x, y, label='Standard Normal Distribution')
    plt.fill_between(x, y, where=(x <= -z_critical), color='red', alpha=0.5, label=f'Type I Error (α = {type_1_error:.3f})')
    plt.fill_between(x, y, where=(x >= z_critical), color='red', alpha=0.5)
    plt.fill_between(x, y, where=(x >= -z_critical) & (x <= z_critical), color='green', alpha=0.5, label=f'Type II Error (β ≈ {type_2_error:.3f})')
    plt.axvline(x=z_stat, color='blue', linestyle='--', label=f'Test Statistic (Z = {z_stat:.2f})')
    plt.axvline(x=z_critical, color='black', linestyle='--', label=f'Critical Value (Z = ±{z_critical:.2f})')
    plt.axvline(x=-z_critical, color='black', linestyle='--')
    plt.title('Type I and Type II Errors in Hypothesis Testing')
    plt.xlabel('Z-Score')
    plt.ylabel('Probability Density')
    plt.legend()
    plt.grid(True)
    plt.show()

    return z_stat, type_1_error, type_2_error
```

```python
#9.Write a Python function to calculate the confidence interval for a sample mean and explain its significance.
import numpy as np
from scipy import stats

def calculate_confidence_interval(data, confidence=0.95):
    """
    Calculate the confidence interval for a sample mean.

    Parameters:
    - data: array-like, sample data
    - confidence: float, confidence level (default is 0.95)

    Returns:
    - tuple: (lower bound, upper bound) of the confidence interval
    """
    # Sample size
    n = len(data)

    # Sample mean
    mean = np.mean(data)

    # Sample standard deviation (ddof=1 for an unbiased estimator)
    std_dev = np.std(data, ddof=1)

    # Standard error of the mean
    sem = std_dev / np.sqrt(n)

    # Degrees of freedom
    dof = n - 1

    # t-critical value for the confidence level
    t_critical = stats.t.ppf((1 + confidence) / 2, dof)

    # Margin of error
    margin_of_error = t_critical * sem

    # Confidence interval
    ci_lower = mean - margin_of_error
    ci_upper = mean + margin_of_error

    return ci_lower, ci_upper
```

```python
#10.Write a Python program to calculate the margin of error for a given confidence level using sample data.
import numpy as np
from scipy.stats import norm

def calculate_margin_of_error(population_std, sample_size, confidence_level=0.95):
    """
    Calculate the margin of error for a sample mean.

    Parameters:
    - population_std: float, population standard deviation
    - sample_size: int, sample size
    - confidence_level: float, desired confidence level (default is 0.95)

    Returns:
    - float: margin of error
    """
    # Calculate the Z-score for the given confidence level
    z_score = norm.ppf(1 - (1 - confidence_level) / 2)

    # Calculate the standard error
    standard_error = population_std / np.sqrt(sample_size)

    # Calculate the margin of error
    margin_of_error = z_score * standard_error

    return margin_of_error
```

```python
#11.Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process.
import scipy.stats as stats

def bayesian_inference(prior, flips, heads):
    # Likelihood: Probability of observing 'heads' heads in 'flips' flips
    likelihood = stats.binom.pmf(heads, flips, 0.5)

    # Evidence: Total probability of observing the evidence under all hypotheses
    evidence = likelihood * prior + (1 - likelihood) * (1 - prior)

    # Posterior: Updated belief after observing the evidence
    posterior = (likelihood * prior) / evidence

    return posterior

# Prior belief that the coin is fair
prior = 0.5

# Observed data: 7 heads in 10 flips
flips = 10
heads = 7

# Compute the posterior probability
posterior = bayesian_inference(prior, flips, heads)
print(f"Posterior probability that the coin is fair: {posterior:.4f}")
```

```python
#12.Perform a Chi-square test for independence between two categorical variables in Python.
#Step 1: Import Required Libraries
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

#Step 2: Create the Contingency Table
# Creating the contingency table
data = np.array([[25, 20, 15],   # Male preferences
                 [30, 10, 20]])  # Female preferences

# Converting to DataFrame for better visualization
df = pd.DataFrame(data, columns=['Prefer_A', 'Prefer_B', 'Prefer_C'], index=['Male', 'Female'])

print("Contingency Table:")
print(df)

#Step 3: Perform the Chi-Square Test
# Performing Chi-Square Test
chi2_stat, p_value, dof, expected = chi2_contingency(data)

# Displaying results
print(f"\nChi-Square Statistic: {chi2_stat:.4f}")
print(f"Degrees of Freedom: {dof}")
print("Expected Frequencies Table:\n", expected)
print(f"P-value: {p_value:.4f}")

#Step 4: Interpret the Results
# Significance level
alpha = 0.05

# Decision
if p_value <= alpha:
    print("\nReject Null Hypothesis: There is a significant association between gender and brand preference.")
else:
    print("\nFail to Reject Null Hypothesis: No significant association between gender and brand preference.")
```

```python
#13. Write a Python program to calculate the expected frequencies for a Chi-square test based on observed data.
import numpy as np
import pandas as pd

# Sample observed data (contingency table)
data = np.array([[25, 20, 15],   # Male preferences
                 [30, 10, 20]])  # Female preferences

# Convert to DataFrame for better visualization
df = pd.DataFrame(data, columns=['Prefer_A', 'Prefer_B', 'Prefer_C'], index=['Male', 'Female'])

# Calculate row totals, column totals, and grand total
row_totals = df.sum(axis=1).values.reshape(-1, 1)
col_totals = df.sum(axis=0).values.reshape(1, -1)
grand_total = df.sum().values

# Calculate expected frequencies
expected = (row_totals @ col_totals) / grand_total

# Convert to DataFrame for better visualization
expected_df = pd.DataFrame(expected, columns=df.columns, index=df.index)

print("Observed Frequencies:")
print(df)
print("\nExpected Frequencies:")
print(expected_df)
```
